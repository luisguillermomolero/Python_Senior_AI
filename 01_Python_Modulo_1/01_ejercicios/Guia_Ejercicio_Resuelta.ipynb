### Ejercicio 1 — Descomposición de billetes

### Ejercicio 2 — Cálculo del IMC

**Enunciado:** Crea un programa que solicite el peso (kg) y la altura (m) de una persona, valide que sean números positivos, calcule el Índice de Masa Corporal (IMC) y muestre el valor con dos decimales junto con la clasificación según los intervalos pedidos.

**Objetivo:** Validación de entradas, uso de operaciones aritméticas y condicionales para clasificación de resultados.

```python
# Pedir peso válido
while True:
    entrada = input("Ingrese su peso en kilogramos (ej. 70.5): ")
    try:
        peso = float(entrada)
        if peso <= 0:
            print("El peso debe ser mayor que 0.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Ingrese un número.")

# Pedir altura válida
while True:
    entrada = input("Ingrese su altura en metros (ej. 1.70): ")
    try:
        altura = float(entrada)
        if altura <= 0:
            print("La altura debe ser mayor que 0.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Ingrese un número.")

# Validación razonable de altura
if altura < 0.5:
    print("Atención: la altura ingresada es muy baja. Verifique la unidad (metros).")

# Cálculo IMC
imc = peso / (altura * altura)
print("IMC calculado: {:.2f}".format(imc))

# Clasificación
if imc < 18.5:
    print("Clasificación: Bajo peso")
elif imc < 25.0:
    print("Clasificación: Normal")
elif imc < 30.0:
    print("Clasificación: Sobrepeso")
else:
    print("Clasificación: Obesidad")
```

### Ejercicio 11 — Validación y clasificación de triángulo

**Enunciado:** Crea un programa que lea tres números diferentes y determine si pueden formar los lados de un triángulo. Si es posible, indicar si es equilátero, isósceles o escaleno. Use la desigualdad triangular para la validación.

**Objetivo:** Condicionales anidados y operadores lógicos para validar relaciones geométricas entre números.

```python
# Lectura validada de tres lados (positivos)
count = 0
while count < 3:
    try:
        if count == 0:
            a = float(input("Ingrese lado a: "))
        elif count == 1:
            b = float(input("Ingrese lado b: "))
        else:
            c = float(input("Ingrese lado c: "))
        if a <= 0 or b <= 0 or c <= 0:
            print("Los lados deben ser mayores que cero.")
            count = 0
            continue
        count += 1
    except ValueError:
        print("Entrada inválida. Ingrese números.")

# Validar desigualdad triangular
if (a + b > c) and (a + c > b) and (b + c > a):
    # Clasificación
    if (a == b) and (b == c):
        print("Es un triángulo equilátero.")
    elif (a == b) or (a == c) or (b == c):
        print("Es un triángulo isósceles.")
    else:
        print("Es un triángulo escaleno.")
else:
    print("No pueden formar un triángulo.")
```

```python
# Validación de entradas
while True:
    try:
        horas = float(input("Ingrese horas trabajadas: "))
        valor_hora = float(input("Ingrese el valor por hora: "))
        if horas < 0 or valor_hora < 0:
            print("Valores no pueden ser negativos.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Ingrese números.")

if horas <= 40:
    salario = horas * valor_hora
else:
    extras = horas - 40
    salario = (40 * valor_hora) + (extras * valor_hora * 1.5)

print("Salario a pagar: ${:.2f}".format(salario))
```

### Ejercicio 13 — Número primo

**Enunciado:** Elabora un programa que solicite un número entero positivo y determine si es primo (divisible solo por 1 y por sí mismo).

**Objetivo:** Uso de ciclos y operador módulo para comprobar divisibilidad y optimizar la búsqueda de factores.

```python
# Validar entero positivo
while True:
    try:
        n = int(input("Ingrese un número entero positivo: "))
        if n <= 0:
            print("Ingrese un número mayor que cero.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Ingrese un número entero.")

if n == 1:
    print("1 no es primo ni compuesto.")
else:
    es_primo = True
    i = 2
    while i * i <= n:
        if n % i == 0:
            es_primo = False
            break
        i = i + 1
    if es_primo:
        print("Es un número primo.")
    else:
        print("No es un número primo.")
```

### Ejercicio 14 — Serie de Fibonacci (n términos)

**Enunciado:** Diseña un programa que muestre los primeros n términos de la serie de Fibonacci, donde n es un entero positivo ingresado por el usuario.

**Objetivo:** Manejo de ciclos y variables acumuladoras para generar secuencias recursivas iterativas.

```python
# Validar n
while True:
    try:
        n = int(input("Ingrese la cantidad de términos (entero positivo): "))
        if n <= 0:
            print("Ingrese un número mayor que cero.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Ingrese un entero.")

# Generar términos sin usar funciones
term = 1
prev = 0
count = 0
while count < n:
    print(term)
    temp = term
    term = prev + term
    prev = temp
    count = count + 1
```

### Ejercicio 26 — Mínimo Común Múltiplo por sumas sucesivas

**Enunciado:** Diseña un programa que calcule el MCM de dos enteros positivos mediante sumas sucesivas (evitar uso de funciones avanzadas).

**Objetivo:** Comprensión de iteración hasta encontrar la primera coincidencia en múltiplos.

```python
# Validar a y b positivos
while True:
    try:
        a = int(input("Ingrese el primer entero positivo a: "))
        b = int(input("Ingrese el segundo entero positivo b: "))
        if a <= 0 or b <= 0:
            print("Ingrese valores positivos.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Ingrese enteros positivos.")

# Sumas sucesivas
multiplo_a = a
multiplo_b = b
while multiplo_a != multiplo_b:
    if multiplo_a < multiplo_b:
        multiplo_a = multiplo_a + a
    else:
        multiplo_b = multiplo_b + b

print("El MCM es:", multiplo_a)
```

### Ejercicio 27 — MCD (algoritmo de Euclides por restas)

**Enunciado:** Crea un programa que calcule el máximo común divisor (MCD) de dos enteros positivos usando restas sucesivas.

**Objetivo:** Aplicación de algoritmo clásico mediante bucles y condicionales para encontrar el máximo común divisor.

```python
# Validar a y b positivos
while True:
    try:
        a = int(input("Ingrese el primer entero positivo a: "))
        b = int(input("Ingrese el segundo entero positivo b: "))
        if a <= 0 or b <= 0:
            print("Ingrese valores positivos.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Ingrese enteros positivos.")

# Algoritmo por restas sucesivas
x = a
y = b
while x != y:
    if x > y:
        x = x - y
    else:
        y = y - x

print("El MCD es:", x)
```

**Enunciado:** Desarrolla un programa que genere un patrón numérico en forma de triángulo rectángulo. El usuario ingresa n y se imprime líneas del 1 hasta n repitiendo el número según la línea.

```python
# Validar n
while True:
    try:
        n = int(input("Ingrese un entero n (>=1) para construir el patrón: "))
        if n < 1:
            print("Ingrese un entero mayor o igual a 1.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Ingrese un entero.")

i = 1
while i <= n:
    # Imprimir la línea i repitiendo el dígito i, usando un contador interno
    j = 0
    linea = ""
    while j < i:
        linea = linea + str(i)
        j = j + 1
    print(linea)
    i = i + 1
```

### Ejercicio 29 — Número Armstrong

**Enunciado:** Escribe un programa que determine si un número entero es un número Armstrong (la suma de sus dígitos elevados al número de dígitos es igual al número).

**Objetivo:** Manipulación numérica avanzada: conteo de dígitos, extracción y uso de potencias.

```python
# Validar entero positivo
while True:
    try:
        num = int(input("Ingrese un número entero positivo: "))
        if num < 0:
            print("Ingrese un número no negativo.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Ingrese un entero.")

# Contar dígitos
temp = num
digitos = 0
if temp == 0:
    digitos = 1
else:
    while temp > 0:
        temp = temp // 10
        digitos = digitos + 1

# Calcular suma de potencias
suma = 0
temp = num
while temp > 0:
    dig = temp % 10
    # elevar dig a la potencia 'digitos' usando multiplicación repetida
    pot = 1
    k = 0
    while k < digitos:
        pot = pot * dig
        k = k + 1
    suma = suma + pot
    temp = temp // 10

if suma == num:
    print(num, "es un número Armstrong.")
else:
    print(num, "no es un número Armstrong.")
```

### Ejercicio 30 — Cálculo de tarifa de parqueadero

**Enunciado:** Crea un programa que calcule el total a pagar en un parqueadero. El usuario ingresa horas de permanencia y tipo de vehículo (carro o moto). Tarifa: carro $5000/h, moto $2000/h. Si la permanencia supera 10 horas, aplicar 10% de descuento sobre el total.

**Objetivo:** Uso combinado de condicionales anidados y aritmética para resolver un problema de negocio.

```python
# Validar horas y tipo
while True:
    try:
        horas = float(input("Ingrese las horas de permanencia (por ejemplo 3.5): "))
        if horas < 0:
            print("Las horas no pueden ser negativas.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Ingrese un número.")

# Pedir tipo (validar con opciones simples)
while True:
    tipo = input("Ingrese tipo de vehículo ('carro' o 'moto'): ").strip().lower()
    if tipo == "carro" or tipo == "moto":
        break
    print("Tipo inválido. Ingrese 'carro' o 'moto'.")

# Calcular tarifa base por hora
if tipo == "carro":
    tarifa = 5000.0
else:
    tarifa = 2000.0

total = horas * tarifa

# Aplicar descuento si corresponde
if horas > 10:
    descuento = total * 0.10
    total = total - descuento
    print("Se aplica descuento del 10% por permanencia mayor a 10 horas. Descuento: ${:.2f}".format(descuento))

print("Total a pagar: ${:.2f}".format(total))
```
